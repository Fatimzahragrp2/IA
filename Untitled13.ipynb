```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12,5)

# =========================
# DONNÉES
# =========================

rendements_A = np.array([
1.2, 0.8, -0.5, 1.5, 0.9, 1.1, 0.7, 1.3, 1.0, 0.6, 1.4, 0.8,
1.1, 0.9, -0.3, 1.2, 1.0, 1.5, 0.8, 1.3, 0.9, 1.1, 1.2, 1.0
])

rendements_B = np.array([
4.5, -2.1, 6.2, -3.5, 5.8, 7.1, -1.8, 4.9, 3.2, -4.2, 8.5, -2.7,
5.1, 6.8, -3.1, 7.3, 4.5, -2.9, 6.7, 5.3, -3.8, 7.9, 4.2, 5.5
])

capital = 500000
rf = 3

# =========================
# QUESTION 1.1 — STATISTIQUES
# =========================

def stats_portefeuille(r):
    mean = np.mean(r)
    std = np.std(r, ddof=1)
    median = np.median(r)
    rendement_annuel = ((1 + mean/100)**12 - 1) * 100
    volatilite_annuelle = std * np.sqrt(12)
    return mean, std, median, rendement_annuel, volatilite_annuelle

stats_A = stats_portefeuille(rendements_A)
stats_B = stats_portefeuille(rendements_B)

print("Portefeuille A :", stats_A)
print("Portefeuille B :", stats_B)

# =========================
# QUESTION 1.2 — VISUALISATION
# =========================

fig, ax = plt.subplots(1,2)

ax[0].hist(rendements_A, alpha=0.6, label="A")
ax[0].hist(rendements_B, alpha=0.6, label="B")
ax[0].legend()
ax[0].set_title("Histogrammes")

ax[1].boxplot([rendements_A, rendements_B], labels=["A","B"])
ax[1].set_title("Boxplots")

plt.show()

# =========================
# QUESTION 1.3 — VaR
# =========================

def var_portefeuille(mean, std, r_annuel, vol_annuel):
    z = -1.65
    var_mensuelle = mean + z*std
    var_annuelle = r_annuel + z*vol_annuel
    var_euros = capital * var_annuelle / 100
    return var_mensuelle, var_annuelle, var_euros

var_A = var_portefeuille(*stats_A[:2], stats_A[3], stats_A[4])
var_B = var_portefeuille(*stats_B[:2], stats_B[3], stats_B[4])

print("VaR A :", var_A)
print("VaR B :", var_B)

# Test normalité
print("Shapiro A :", stats.shapiro(rendements_A))
print("Shapiro B :", stats.shapiro(rendements_B))

# =========================
# QUESTION 1.4 — SHARPE
# =========================

sharpe_A = (stats_A[3] - rf) / stats_A[4]
sharpe_B = (stats_B[3] - rf) / stats_B[4]

print("Sharpe A :", sharpe_A)
print("Sharpe B :", sharpe_B)
```

```python
# =========================
# QUESTION 2.1
# =========================

prior = 0.05
p_retard_def = 0.80
p_retard_non = 0.10

p_retard = p_retard_def*prior + p_retard_non*(1-prior)
posterior = (p_retard_def*prior) / p_retard

print("P(Défaut|Retard) =", posterior)

# =========================
# QUESTION 2.2 — Séquentiel
# =========================

p_dec_def = 0.65
p_dec_non = 0.15

prior2 = posterior
p_dec = p_dec_def*prior2 + p_dec_non*(1-prior2)
posterior2 = (p_dec_def*prior2)/p_dec

print("P(Défaut|Retard et Découvert) =", posterior2)

# Graphique évolution
plt.plot([prior*100, posterior*100, posterior2*100], marker="o")
plt.xticks([0,1,2], ["Prior","Après Retard","Après Découvert"])
plt.ylabel("Probabilité défaut (%)")
plt.show()

# =========================
# QUESTION 2.3 — Fonction générique
# =========================

def bayes_update(prior, likelihood_pos, likelihood_neg):
    p_evidence = likelihood_pos*prior + likelihood_neg*(1-prior)
    return (likelihood_pos*prior)/p_evidence

print(bayes_update(0.15, 0.80, 0.10))

# =========================
# QUESTION 2.4 — Precision
# =========================

tp = 400
fp = 950

precision = tp/(tp+fp)
print("Precision :", precision)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc

# =========================
# GÉNÉRATION DATASET
# =========================

np.random.seed(42)
n = 2000

age = np.random.randint(25,66,n)
salaire = np.random.normal(50000,20000,n).clip(20000,120000)
anciennete = np.random.exponential(5,n).clip(0,30)
dette = np.random.normal(25000,15000,n).clip(0,80000)
ratio = dette/salaire
nb_credits = np.random.poisson(1.5,n).clip(0,5)
retards = np.random.poisson(2,n).clip(0,10)
score = np.random.normal(650,100,n).clip(300,850)

proba = (0.05 +
         0.15*(ratio>0.5) +
         0.10*(retards>3) +
         0.08*(score<600) +
         0.05*(nb_credits>2)).clip(0,0.85)

defaut = (np.random.rand(n) < proba).astype(int)

df = pd.DataFrame({
    "age":age,
    "salaire":salaire,
    "anciennete":anciennete,
    "dette":dette,
    "ratio":ratio,
    "nb_credits":nb_credits,
    "retards":retards,
    "score":score,
    "defaut":defaut
})

print(df.head())
print("Taux défaut :", df["defaut"].mean())
print(df["defaut"].value_counts())

# =========================
# SPLIT + SCALING
# =========================

X = df.drop("defaut", axis=1)
y = df["defaut"]

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# =========================
# OPTIMISATION K
# =========================

scores = []
for k in range(1,21):
    model = KNeighborsClassifier(n_neighbors=k)
    score = cross_val_score(model,X_train,y_train,cv=5,scoring="recall").mean()
    scores.append(score)

best_k = np.argmax(scores)+1
print("Meilleur K :", best_k)

# =========================
# MODÈLE FINAL
# =========================

model = KNeighborsClassifier(n_neighbors=best_k)
model.fit(X_train,y_train)

y_pred = model.predict(X_test)

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))

# =========================
# ROC / AUC
# =========================

y_prob = model.predict_proba(X_test)[:,1]
fpr, tpr, _ = roc_curve(y_test,y_prob)
roc_auc = auc(fpr,tpr)

plt.plot(fpr,tpr,label="AUC="+str(round(roc_auc,3)))
plt.plot([0,1],[0,1],'--')
plt.legend()
plt.show()

# =========================
# ROI MÉTIER
# =========================

tn, fp, fn, tp = confusion_matrix(y_test,y_pred).ravel()

roi = (tp*15000) - (fn*15000) - (fp*500) - (fp*1200)

print("ROI estimé :", roi, "€")
```
